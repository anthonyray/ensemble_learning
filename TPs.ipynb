```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor

max_depth = 1
# Create a random dataset
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 1 * (0.5 - rng.rand(16))
y = y.reshape(80,1)

# Bootstraping

n = X.shape[0]
random_indices = np.random.randint(0,n,(n,))

XX = X[random_indices]
yy = y[random_indices]

for i in range(1,L):
    random_indices = np.random.randint(0,n,(n,))
    random_sample = X[random_indices]
    random_labels = y[random_indices]
    XX = np.hstack((XX,random_sample))
    yy = np.hstack((yy,random_labels))
    
stumps = [DecisionTreeRegressor(max_depth=max_depth) for i in range(L)]

"""
Training stumps
"""

for i,stump in enumerate(stumps):
    stump.fit(XX[:,i].reshape((n,1)),yy[:,i].reshape((n,1)))

"""
Aggregating stumps
"""
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]

y_pred = np.zeros((X_test.shape[0],))
for index,stump in enumerate(stumps):
    y_pred = y_pred + stump.predict(X_test)
    
y_pred = y_pred / L    
```

# Random Forests

Dans la suite, on évalue le score par 5-folds cross-validation des Random-Forests sur les datasets boston, diabetes, iris
et digits.

```python
from sklearn import datasets
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.svm import SVR

from sklearn import cross_validation

iris = datasets.load_iris()
boston = datasets.load_boston()
diabetes = datasets.load_diabetes()
digits = datasets.load_digits()


clf = RandomForestClassifier()
scores = cross_validation.cross_val_score(clf, iris.data, iris.target, cv=5)
print("Accuracy of Random Forests for IRIS : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = RandomForestClassifier()
scores = cross_validation.cross_val_score(clf, digits.data, digits.target, cv=5)
print("Accuracy of Random Forests for digits : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = RandomForestRegressor()
scores = cross_validation.cross_val_score(clf, boston.data, boston.target, cv=5)
print("Accuracy of Random Forests for boston : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = RandomForestRegressor()
scores = cross_validation.cross_val_score(clf, diabetes.data, diabetes.target, cv=5)
print("Accuracy of Random Forests for diabetes : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = SVC(kernel='linear')
scores = cross_validation.cross_val_score(clf, iris.data, iris.target, cv=5)
print("Accuracy of SVM for IRIS : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = SVC(kernel='linear')
scores = cross_validation.cross_val_score(clf, digits.data, digits.target, cv=5)
print("Accuracy of SVM for digits : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = SVR(kernel='linear')
scores = cross_validation.cross_val_score(clf, boston.data, boston.target, cv=5)
print("Accuracy of SVR for boston : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = SVR(kernel='linear')
scores = cross_validation.cross_val_score(clf, diabetes.data, diabetes.target, cv=5)
print("Accuracy of SVR for diabetes : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

#### Question 4.1 

On peut décomposer l'expression de l'espérance conditionelle. 
\begin{align}
\\E \Bigg[ e^{-yf(x)} \rvert  X = x \Bigg] = P\Big( Y = 1 \rvert X = x\Big)e^{-f(x)} + P\Big( Y = -1 \rvert X = x\Big)e^{f(x)} \\
\end{align}

On cherche à minimiser pour **f** l'expression ci-dessus. Dérivons l'espérance conditionelle par rapport à f : 
\begin{align}
\frac{\partial E \Big[ e^{-yF(x)} \rvert  X = x \Big]}{\partial f(x)} = - P\Big( Y = 1 \rvert X = x\Big)e^{-f(x)} + P\Big( Y = -1 \rvert X = x\Big)e^{f(x)}
\end{align}

Avec la notation définie précédemment : 
\begin{align}
\\P \Big( Y = 1 \rvert  X = x \Big) = \eta(x)
\end{align}

On a : 
\begin{align}
\frac{\partial E \Big[ e^{-yF(x)} \rvert  X = x \Big]}{\partial f(x)} = - \eta (x)e^{-f(x)} + (1-\eta (x))e^{f(x)} = 0
\end{align}

d'où 
\begin{align}
\\f^{*}(x) = \frac{1}{2}log\frac{\eta (x)}{1 - \eta (x)}
\end{align}

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_digits
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import zero_one_loss

depths = [1,2,10]

for max_depth in depths:
    n_estimators = 100
    digits = load_digits()
    X,y = digits.data, digits.target

    stump = DecisionTreeClassifier(max_depth=max_depth)
    stump.fit(X,y)

    ada = AdaBoostClassifier(base_estimator=stump)
    ada.fit(X,y)

    print "Score with Adaboost for a decision tree with a depth of " + str(max_depth) + " : " + str(ada.score(X,y))

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import zero_one_loss

n_estimators = 120
max_depth = 10
digits = datasets.load_digits()
X,y = digits.data, digits.target

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

decision_tree = DecisionTreeClassifier(max_depth=max_depth)
decision_tree.fit(X_train,y_train)

ada_discrete = AdaBoostClassifier(base_estimator=decision_tree)
ada_discrete.fit(X_train, y_train)

ada_discrete_err = np.zeros((n_estimators,))
for i, y_pred in enumerate(ada_discrete.staged_predict(X_test)):
    ada_discrete_err[i] = zero_one_loss(y_pred, y_test)

ada_discrete_err_train = np.zeros((n_estimators,))
for i, y_pred in enumerate(ada_discrete.staged_predict(X_train)):
    ada_discrete_err_train[i] = zero_one_loss(y_pred, y_train)

fig = plt.figure(figsize=(15,6))
plt.plot(np.arange(n_estimators) + 1, ada_discrete_err,
        label='Erreur test',
        color='red')
plt.plot(np.arange(n_estimators) + 1, ada_discrete_err_train,
        label="Erreur d'aprentissage ",
        color='blue')
plt.legend()
```
