### Question 2

Dans la suite, nous mettons en place un script mettant en oeuvre le bagging avec des arbres de profondeur 1.

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor

max_depth = 1
# Create a random dataset
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 1 * (0.5 - rng.rand(16))
y = y.reshape(80,1)

# Bootstraping

n = X.shape[0]
random_indices = np.random.randint(0,n,(n,))

XX = X[random_indices]
yy = y[random_indices]

for i in range(1,L):
    random_indices = np.random.randint(0,n,(n,))
    random_sample = X[random_indices]
    random_labels = y[random_indices]
    XX = np.hstack((XX,random_sample))
    yy = np.hstack((yy,random_labels))
    
stumps = [DecisionTreeRegressor(max_depth=max_depth) for i in range(L)]

"""
Training stumps
"""

for i,stump in enumerate(stumps):
    stump.fit(XX[:,i].reshape((n,1)),yy[:,i].reshape((n,1)))

"""
Aggregating stumps
"""
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]

y_pred = np.zeros((X_test.shape[0],))
for index,stump in enumerate(stumps):
    y_pred = y_pred + stump.predict(X_test)
    
y_pred = y_pred / L    
```

```python
X = np.sort(5 * rng.rand(80, 1), axis=0)
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]

def prediction_for_given_noise_level(noise_lvl):
    L = 25
    y = np.sin(X).ravel()
    y[::5] += noise_lvl * (0.5 - rng.rand(16))
    y = y.reshape(80,1)
    
    # Bootstraping
    
    n = X.shape[0]
    random_indices = np.random.randint(0,n,(n,))
    
    XX = X[random_indices]
    yy = y[random_indices]
    
    for i in range(1,L):
        random_indices = np.random.randint(0,n,(n,))
        random_sample = X[random_indices]
        random_labels = y[random_indices]
        XX = np.hstack((XX,random_sample))
        yy = np.hstack((yy,random_labels))
        
    stumps = [DecisionTreeRegressor(max_depth=5) for i in range(L)]
    
    """
    Training stumps
    """
    
    for i,stump in enumerate(stumps):
        stump.fit(XX[:,i].reshape((n,1)),yy[:,i].reshape((n,1)))
    
    """
    Aggregating stumps
    """
    y_pred = np.zeros((X_test.shape[0],))
    for index,stump in enumerate(stumps):
        y_pred = y_pred + stump.predict(X_test)
        
    y_pred = y_pred / L
    return (y,y_pred)

predictions = [prediction_for_given_noise_level(i) for i in [0.001,0.01,0.1,1,10,20]]

fig, axes = plt.subplots(6,1, figsize=(8,12))

for i,ax in enumerate(axes):
    ax.scatter(X,predictions[i][0],c="k",label="data")
    ax.plot(X_test, predictions[i][1], c="g", label="L : %d" % (i+1))
    ax.legend()
```

# Random Forests

Dans la suite, on évalue le score par 5-folds cross-validation des Random-Forests sur les datasets boston, diabetes, iris
et digits.

```python
from sklearn import datasets
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.svm import SVR

from sklearn import cross_validation

iris = datasets.load_iris()
boston = datasets.load_boston()
diabetes = datasets.load_diabetes()
digits = datasets.load_digits()


clf = RandomForestClassifier()
scores = cross_validation.cross_val_score(clf, iris.data, iris.target, cv=5)
print("Accuracy of Random Forests for IRIS : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = RandomForestClassifier()
scores = cross_validation.cross_val_score(clf, digits.data, digits.target, cv=5)
print("Accuracy of Random Forests for digits : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = RandomForestRegressor()
scores = cross_validation.cross_val_score(clf, boston.data, boston.target, cv=5)
print("Accuracy of Random Forests for boston : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = RandomForestRegressor()
scores = cross_validation.cross_val_score(clf, diabetes.data, diabetes.target, cv=5)
print("Accuracy of Random Forests for diabetes : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = SVC(kernel='linear')
scores = cross_validation.cross_val_score(clf, iris.data, iris.target, cv=5)
print("Accuracy of SVM for IRIS : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = SVC(kernel='linear')
scores = cross_validation.cross_val_score(clf, digits.data, digits.target, cv=5)
print("Accuracy of SVM for digits : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = SVR(kernel='linear')
scores = cross_validation.cross_val_score(clf, boston.data, boston.target, cv=5)
print("Accuracy of SVR for boston : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

clf = SVR(kernel='linear')
scores = cross_validation.cross_val_score(clf, diabetes.data, diabetes.target, cv=5)
print("Accuracy of SVR for diabetes : %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

Affichons la probabilité de prédiction des classes en fonction du nombre d'estimateurs de la fôret aléatoire. 

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import shuffle
# Parameters
n_estimators = 2
plot_colors = "bry"
plot_step = 0.02
# Load data
iris = load_iris()
X, y = iris.data[:, :2], iris.target
model = RandomForestClassifier(n_estimators=n_estimators)

# Shuffle
X, y = shuffle(X, y, random_state=42)
# Standardize
mean, std = X.mean(axis=0), X.std(axis=0)
X = (X - mean) / std
# Train
clf = model.fit(X, y)
# Plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
np.arange(y_min, y_max, plot_step))
plt.close('all')
for tree in model.estimators_:
    Z = tree.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, alpha=1. / n_estimators, cmap=plt.cm.Paired)
plt.axis("tight")
# Plot the training points
for i, c in zip(xrange(3), plot_colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=c, label=iris.target_names[i],
    cmap=plt.cm.Paired)
plt.show()
```
